```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

def generate_code_with_gpt2(prompt, max_length=50):
    tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
    model = GPT2LMHeadModel.from_pretrained("gpt2")

    inputs = tokenizer.encode(prompt, return_tensors='pt')
    outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1)

    return tokenizer.decode(outputs[0], skip_special_tokens=True)

generated_code_list = []

for query_id in range(1, 10):
    prompt = query_records[query_id - 1]['text']  # 使用查询文本作为prompt
    generated_code = generate_code_with_gpt2(prompt)

    # 将生成的代码存储到数组中
    generated_code_list.append({'query_id': str(query_id), 'generated_code': generated_code})

# 打印生成的代码
for generated_record in generated_code_list:
    print(f"Query ID: {generated_record['query_id']}")
    print(f"Generated Code:\n{generated_record['generated_code']}")
    print("\n")
```

```python
import numpy as np

def find_optimal_weights(answer, answer2):
    best_weighted_average1 = None
    best_weighted_average2 = None
    best_rmse1 = float("inf")
    best_rmse2 = float("inf")
    best_i1 = None
    best_i2 = None

    for i in np.arange(0, 1, 0.01):
        # 计算第一组的加权平均
        mid = len(answer) // 2
        weighted_average1 = [(i * a2) + ((1 - i) * a1) for a1, a2 in zip(answer[:mid], answer2[:mid])]
        rmse1 = calculate_rmse(weighted_average1)

        # 计算第二组的加权平均
        weighted_average2 = [(i * a2) + ((1 - i) * a1) for a1, a2 in zip(answer[mid:], answer2[mid:])]
        rmse2 = calculate_rmse(weighted_average2)

        # 检查是否找到更小的 RMSE 和对应的权重
        if rmse1 < best_rmse1:
            best_weighted_average1 = weighted_average1
            best_rmse1 = rmse1
            best_i1 = i

        if rmse2 < best_rmse2:
            best_weighted_average2 = weighted_average2
            best_rmse2 = rmse2
            best_i2 = i

    return best_weighted_average1, best_weighted_average2, best_rmse1, best_rmse2, best_i1, best_i2

# 调用函数来查找每组的最佳权重
optimal_weighted_average_1, optimal_weighted_average_2, min_rmse_1, min_rmse_2, best_i_1, best_i_2 = find_optimal_weights(answer, answer2)

# 打印结果
print("最佳权重和最小 RMSE（组1）：", optimal_weighted_average_1)
print("最佳权重和最小 RMSE（组2）：", optimal_weighted_average_2)
print("最小 RMSE 值（组1）：", min_rmse_1)
print("最小 RMSE 值（组2）：", min_rmse_2)
print("对应的最佳权重 i（组1）：", best_i_1)
print("对应的最佳权重 i（组2）：", best_i_2)
```
